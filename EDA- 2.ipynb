#### Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset to identify the peak travel season. What features would you analyze to identify the peak season, and how would you present your findings to your boss?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df=pd.read_excel('flight_price.xlsx')
# Convert the date column to datetime format
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])
# Extract the month and day from the 'Date_of_Journey' column
df['Month'] = df['Date_of_Journey'].dt.month
df['Day'] = df['Date_of_Journey'].dt.day
# Group the data by month and calculate the average flight price for each month
monthly_avg_price = df.groupby('Month')['Price'].mean()
# Visualize the average flight prices for each month
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(monthly_avg_price.index, monthly_avg_price.values, marker='o')
plt.xlabel('Month')
plt.ylabel('Average Flight Price')
plt.title('Average Flight Price by Month')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
# Convert 'Date_of_Journey' column to datetime format
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])

# Extract month from the date
df['Month'] =df['Date_of_Journey'].dt.month

# Map month to season
season_mapping = {
    12: 'Winter', 1: 'Winter', 2: 'Winter',
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Autumn', 10: 'Autumn', 11: 'Autumn'
}
df['Season'] = df['Month'].map(season_mapping)
```

```python
# Group by season and calculate average flight price
season_price = df.groupby('Season')['Price'].mean()

# Display peak season
peak_season = season_price.idxmax()
print("Peak Season:", peak_season)
```

Missing values can impact the analysis in various ways:

Bias: If missing values are not handled properly, it can introduce bias in the analysis and modeling. For instance, if certain types of apps tend to have more missing values in a particular column, it can lead to biased insights about those apps.

Reduced Sample Size: Missing values can reduce the effective sample size for analysis, which might lead to less accurate results.

Model Performance: Missing values can affect the performance of machine learning models as many algorithms do not handle missing values well.

Incorrect Insights: If not handled properly, missing values can distort the relationships between variables and lead to incorrect insights.


#### Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Playstore dataset
url = "https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv"
data = pd.read_csv(url)

# Select the relevant columns
data_subset = data[['Rating', 'Size']]

# Drop rows with missing values
data_subset.dropna(inplace=True)

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data_subset['Size'], data_subset['Rating'], alpha=0.5)
plt.xlabel('Size')
plt.ylabel('Rating')
plt.title('Scatter Plot of App Size vs. Rating')
plt.grid(True)
plt.show()
```

#### Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Playstore dataset
url = "https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv"
data = pd.read_csv(url)

# Clean 'Price' column and convert to numeric
data['Price'] = data['Price'].apply(lambda x: x.replace('$', '') if '$' in str(x) else x)
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

# Group by app type and calculate average price
avg_price_by_type = data.groupby('Type')['Price'].mean()

# Create a bar chart
plt.figure(figsize=(10, 6))
avg_price_by_type.plot(kind='bar', color='blue')
plt.title("Average Prices by App Type")
plt.xlabel("App Type")
plt.ylabel("Average Price")
plt.xticks(rotation=0)
plt.show()
```

#### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
import pandas as pd

# Load the Google Playstore dataset
url = "https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv"
data = pd.read_csv(url)

# Create a frequency table of app names and installs
app_install_freq = data.groupby('App')['Installs'].sum().sort_values(ascending=False)

# Get the top 10 most popular apps
top_10_apps = app_install_freq.head(10)

# Display the top 10 most popular apps
print("Top 10 Most Popular Apps by Installs:")
print(top_10_apps)
```

```python
import pandas as pd

# Load the Google Playstore dataset
url = "https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv"
data = pd.read_csv(url)

# Count the number of apps in each category
category_counts = data['Category'].value_counts()

# Display the top 10 most popular app categories
top_categories = category_counts.head(10)
print("Top 10 Most Popular App Categories:")
print(top_categories)

import matplotlib.pyplot as plt

# Plot the top 10 app categories
plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar', color='skyblue')
plt.xlabel("App Category")
plt.ylabel("Number of Apps")
plt.title("Top 10 Most Popular App Categories")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
