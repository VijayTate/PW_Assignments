Matplotlib is a popular data visualization library in Python. It provides a wide range of tools and functions to create various types of plots and charts, allowing users to effectively visualize and present their data.

@ Data Visualization: Matplotlib helps in creating informative and visually appealing plots, charts, and graphs to represent data in a meaningful way.

@ Model Evaluation: It helps in visualizing the performance and evaluation metrics of machine learning and statistical models.

@ Line Plot: Used to visualize the relationship between two continuous variables over a continuous interval.

@ Bar Plot: Used to compare categorical data by representing it as rectangular bars.

@ Histogram: Used to visualize the distribution of a single variable by dividing it into bins and showing the frequency or count in each bin.

@ Scatter Plot: Used to explore the relationship between two continuous variables, often used to identify patterns or correlations.

@ Pie Chart: Used to represent data as a circular chart, where each category is represented by a slice of the pie based on its proportion.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.title('This is title')
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
    
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple smaller plots, each with its own axes and data visualization.

The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the subplot you want to create. The index ranges from 1 to the total number of subplots in the grid.

Here are some reasons why the subplot() function is useful:

(1) Multiple Plots: With subplot(), you can create multiple plots within a single figure, making it easier to compare and visualize different aspects of the data side by side.

(2) Grid Layout: The function allows you to define a grid layout for subplots, specifying the number of rows and columns. This helps in organizing and arranging the subplots in a structured manner.

(3) Flexibility: You can customize the size, position, and spacing of the subplots in the grid using subplot(). This allows you to create complex layouts and adjust the visual representation as needed.

(4) Code Efficiency: By using subplot(), you can avoid creating multiple separate figure objects and manage the layout of subplots more efficiently in a single figure.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Line 2")

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Line 3")

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Line 4")


plt.tight_layout()
plt.show()
```

A bar plot, also known as a bar chart, is a type of visualization that represents categorical data with rectangular bars. It is used to compare and display the values of different categories or groups.

Bar plots are commonly used to:

(1) Compare data: Bar plots make it easy to compare the values of different categories or groups. The length of each bar represents the value or quantity associated with that category, allowing for quick visual comparison.

(2) Show distribution: Bar plots can be used to show the distribution of data across different categories. For example, you can use a bar plot to visualize the frequency of occurrence of different categories.

(3) Present trends or patterns: Bar plots can reveal trends or patterns in the data. They can be used to show how the values of different categories change over time or across different conditions.

(4) Highlight relationships: Bar plots can also be used to highlight relationships between different categories or groups. For example, you can use grouped bar plots to compare the values of multiple variables for each category.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company,profit)
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of a set of numerical data through its quartiles and outliers. It is used to understand the distribution and variability of the data, identify outliers, and compare data between different categories or groups.

Box plots are commonly used to:

(1) Identify central tendency: The position of the median within the box provides information about the central tendency of the data. The box itself represents the range within which the middle 50% of the data falls.

(2) Analyze spread and variability: The length of the box and whiskers gives an indication of the spread and variability of the data. A longer box or wider whiskers suggest a larger spread or variability.

(3) Compare data between groups: Box plots can be used to compare data between different categories or groups. By plotting multiple boxes side by side, it becomes easy to compare their medians, ranges, and distributions.

(4) Detect outliers: Outliers, which are data points that significantly deviate from the rest of the data, are visually represented as individual points outside the whiskers. Box plots help in identifying and examining these outliers.

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
plt.boxplot(data)
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot')
```
